```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([5, 12])
y = np.array([2, 9])

plt.plot(x, y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,17,27,56]
plt.figure(figsize=(8,5))
plt.plot(x,y,color='maroon',linestyle='-.',marker='d',linewidth=2)
plt.xlabel('krupa')
plt.ylabel('Rupa')
plt.title("line chart")
plt.legend(['my data series'])
plt.grid(True)
plt.show()
```

```python
categories=['kaen','ryukiiii','pavv','chand','deek','bind','cher','sinn']
values=[10,9,6,8,6,4,3,5]
plt.figure(figsize=(8,5))
plt.bar(categories,values,color='gray')
plt.xlabel('names')
plt.ylabel('values')
plt.title('percentage')
plt.show()
```

```python
data = [12, 15, 13, 17, 16, 16, 19, 12, 14, 16, 14, 17, 18, 18, 15, 14]
plt.hist(data, bins=17, edgecolor='purple',color='gold')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('geo')
plt.show()
```

```python
A = [1, 2, 3, 4, 5,6,]
B= [7,8,9,10,11,12]
plt.scatter(x, y, color='brown',marker='d')
plt.xlabel('A Value')
plt.ylabel('B Value')
plt.title('DATA')
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors=['lightblue','pink','lightgreen','gray']
plt.pie(sizes,colors=colors, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Simple Pie Chart Example')
plt.axis('equal')
plt.show()
```

```python
data = [7, 8, 5, 6, 7, 9, 12, 5, 7, 8, 6]
plt.boxplot(data)
plt.title('Simple Box Plot Example')
plt.ylabel('Values')
plt.show()
```

```python
data = np.random.rand(10, 10)
plt.imshow(data, cmap='bone', interpolation='nearest')
plt.colorbar()
plt.title('Simple Heatmap Example')
plt.show()
```
